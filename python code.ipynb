```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()

classifier.add(Dense(input_dim=13,units=30,kernel_initializer='uniform',activation='sigmoid'))

classifier.add(Dense(units=30,kernel_initializer='uniform',activation='sigmoid'))

classifier.add(Dense(units=15,kernel_initializer='uniform',activation='sigmoid'))

classifier.add(Dense(units=8,kernel_initializer='uniform',activation='sigmoid'))

classifier.add(Dense(units=4,kernel_initializer='uniform',activation='sigmoid'))

classifier.add(Dense(units=2,kernel_initializer='uniform',activation='sigmoid'))

classifier.add(Dense(units=1,activation='sigmoid'))

classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

classifier.fit(x_train,y_train,batch_size=300,epochs=10)
```

###### 

```python
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
```
